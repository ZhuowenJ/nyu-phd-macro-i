## Reading Jupyter notebooks

Jupyter is a webbrowser-based environment that allows you to write documents that integrate executable programming code, for example Python, Julia, or R (but lately also Matlab or Stata). This allows you to view LaTeX/HTML formatted text, executable code, and the code output including graphs on a single page.

For example, the document that you are just reading is a Jupyter notebook, and the following block is a piece of code that draws 100 random Normally distributed numbers and plots them as a trajectory. You should see seven lines of code with a graph underneath.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10,6)
ϵ_values = np.random.randn(100)
plt.plot(ϵ_values)
plt.show()
```

The filename of this document is <tt>setup-instructions.ipynb</tt> and it is hosted on GitHub, in the repository <tt>jborovicka/nyu-phd-macro-i</tt>, in subfolder <tt>jupyter</tt>. If you download the file and open it in a text editor, you will learn it is written in JSON file format, and it needs an interpreter that formats the text and runs the included Python code to produce the graph.

When you open the <a href="https://github.com/jborovicka/nyu-phd-macro-i">nyu-phd-macro-i GitHub repository</a>, navigate to the <tt>jupyter</tt> folder and click on <tt>setup-instructions.ipynb</tt>, GitHub formats the Jupyter notebook, interprets the Python code, and produces the graph. In this way, you can read the Jupiter notebooks we provide without ever installing Python.

There are also other services for viewing Jupyter notebooks, for example <a href="https://nbviewer.jupyter.org/">nbviewer</a>. All you need is to provide a URL link to the Jupyter notebook in the search box. For example, you can provide the link to this document, <tt>https<b></b>://github.com/jborovicka/nyu-phd-macro-i/blob/main/jupyter/setup-instructions.ipynb</tt> into the nbviewer search box, and obtain <a href="https://nbviewer.jupyter.org/github/jborovicka/nyu-phd-macro-i/blob/main/jupyter/setup-instructions.ipynb">this page interpreted by nbviewer</a>. Naturally, it looks identical to the one produced by GitHub. This way, you can for example interpret notebooks stored in your Dropbox by generating a Dropbox file URL and entering it into nbviewer.

## Editing Jupyter notebooks online

The above section provides ways how to read interpreted Jupyter notebook without being able to edit them. If you want to work with the notebooks, edit the embedded code and re-run it to produce new output, but you do not want to install Python, you need an online service to do it.

Two such services are <a href="https://mybinder.org/">binder</a> and <a href="https://colab.research.google.com">Google Colaboratory</a>. We integrated both of them into the GitHub repository. When you open the <a href="https://github.com/jborovicka/nyu-phd-macro-i">main page of the repository</a>, you can see two badges, <b>launch binder</b> and <b>Open in Colab</b>.

Clicking on one of the badges opens the whole repository in a virtual server (it will take a bit of time until the binder server is initialized). You can then again open the file jupyter/setup-instructions.ipynb.

Now you can edit the whole notebook. Try changing the argument in <tt>np.random.randn(100)</tt> to <tt>200</tt>, and then press <tt>Ctrl+Enter</tt>. The virtual server will re-generate the plot. Similarly, you can click on any block of the text, then press <tt>Enter</tt>, and edit its contents. Pressing <tt>Ctrl+Enter</tt> again confirms the changes and renders the block.

These operations alter your own local (online) copy of the notebook, not the one stored in the class GitHub repository. Once you are done, you can save your copy of the notebook, for example, Google Colaboratory allows to store it to your Google Drive. You can also create your own copy of the GitHub repository (this is called forking), and work with that. See the GitHub instructions in the <a href="https://python-programming.quantecon.org/getting_started.html">Setting up Your Python Environment</a> of the QuantEcon documentation.

## Installing your own Python environment

The third option is to install your own Python environment. A very useful implementation that incorporates Python and Jupyter notebook functionality is <a href="https://www.anaconda.com/products/individual">Anaconda</a>.

The <a href="https://python-programming.quantecon.org/">Python Programming for Economics and Finance</a> lecture on QuantEcon provides an excellent step-by-step guide on how to install Anaconda and get your hands on Python. If you are a Matlab user, much of the functionality will be quite familiar.

As the lecture explains in detail, Anaconda creates a virtual server on your own computer, to which you connect via your webbrowser. Then you can create, edit, and run Jupyter notebooks (<tt>.ipynb</tt> files) on your own computer through the webbrowser. These <tt>.ipynb</tt> files can be shared as regular files.

While interacting through the webbrowser perhaps takes a bit of time to get used to, the versatility is hugely beneficial. Your work on your local machine looks exactly the same as if you worked remotely on binder or Google Colaboratory, without the need for any additional tools.
